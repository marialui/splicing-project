```python
import pandas as pd
import numpy as np
```

```python
df_denovo=pd.read_csv('DE_novo_splicing_events_bAmy_VS_CTR.txt',sep='\t')
```

```python
# Convert 'coho7' and 'coho8' to integers, handling NaN values
df_denovo['coho7'] = df_denovo['coho7'].fillna(0).astype(int)  # Fill NaN with 0 and convert to int
df_denovo['coho8'] = df_denovo['coho8'].fillna(0).astype(int)  # Fill NaN with 0 and convert to int
```

```python
df_denovo
```

```python
coho=list(df_denovo.columns[1:13])
# Collapse only the specified columns in 'coho' into a single string
df_denovo['collapsed'] = df_denovo[coho].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
# Strip ' nan' from the 'collapsed' column
# Strip ' nan' only at the end of the string in the 'collapsed' column
# Strip multiple occurrences of ' nan' at the end of the string in the 'collapsed' column
df_denovo['collapsed'] = df_denovo['collapsed'].str.replace(r'(\s?0)+$', '', regex=True)

```

```python
df_denovo['DE-NOVO']=False
```

```python
# Assuming df_denovo is already loaded
for classe in df_denovo['Class'].unique():
    novel_path = f'../24_09_2024_DENOVO-output_rMATS_bAmy-vs-CTR/fromGTF.novelSpliceSite.{classe}.txt'
    
    # Read the CSV into a DataFrame
    Ndf = pd.read_csv(novel_path, sep='\t', index_col='ID')
    
    # Collapse values of relevant columns into a single string (use specific columns if needed)
    Ndf['collapsed'] = Ndf.astype(str).agg(' '.join, axis=1)
    
    # Use pd.Series.isin() to check for matching collapsed strings
    matched_indices = df_denovo[df_denovo['collapsed'].isin(Ndf['collapsed'])].index
    
    # Set the 'DE-NOVO' column to True for all matching rows
    df_denovo.loc[matched_indices, 'DE-NOVO'] = True
```

```python
df_denovo=df_denovo.drop(columns=['collapsed'])
df_denovo[['coho7', 'coho8']] = df_denovo[['coho7', 'coho8']].replace(0, np.nan)
df_denovo.to_csv('splicing_events_bAmy_VS_CTR.txt',sep='\t',index=False)
```
