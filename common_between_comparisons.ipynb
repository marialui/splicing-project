```python

# Step 1: Read the two text files into pandas DataFrames
df1 = pd.read_csv('NOT-SIG-splicing_events_bAmy_VS_CTR.txt', sep=',', index_col=0)  # Assuming tab-separated values
df2 = pd.read_csv('NOT-SIG-splicing_events_CBNR20-bAmy-vs-bAmy.txt', sep=',',index_col=0)  # Assuming tab-separated values

# Step 2: Specify the columns to find common rows
common_columns = ['Class', 'GeneID', 'geneSymbol', 'chr', 'strand', 'coho1', 'coho2', 
                  'coho3', 'coho4', 'coho5', 'coho6', 'coho7', 'coho8','DE-NOVO']

# Step 3: Find common rows based on the specified columns
common_df = pd.merge(df1, df2, on=common_columns, how='inner',suffixes=('_bAmy_VS_CTR', '_CBNR20-bAmy-VS-bAmy'))

# Step 4: Save the common rows to a new file (optional)
common_df.to_csv('common_NOT-SIG-splicing_events.txt', sep='\t', index=False)

# Show the resulting common DataFrame
common_df
```
